```julia
using BSplineKit
using DelimitedFiles
using Plots
using Loess
using Interpolations
```

```julia
data = readdlm("AS_0.4.dat")
```

```julia
model = loess(data[:,2], data[:,1], span = 0.2, degree = 2)
```

```julia
x_fine = range(0.18,0.12, length = 200)
y_smooth = predict(model,x_fine)
writedlm("test.dat", [y_smooth x_fine])
```

```julia
plot(x_fine, y_smooth,xlims=[-0.4,0.4],ylims=[0,700])
```
