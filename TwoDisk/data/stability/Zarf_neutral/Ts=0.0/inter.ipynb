```julia
using BSplineKit
using DelimitedFiles
using Plots
using Loess
using Interpolations
```

```julia
data = readdlm("zarf.dat")
```

```julia
p = sortperm(data[:,2])
```

```julia
inte = Interpolations.linear_interpolation(data[p,2],data[p,1])
```

```julia
x_fine = range(-0.278,0.264,length = 2000)
y_smooth = inte.(x_fine)
```

```julia
plot(x_fine, y_smooth,xlims=[-0.4,0.4],ylims=[0,700])
```

```julia
plot(data[p,2], data[p,1],xlims=[-0.4,0.4],ylims=[0,700])
```
