```julia
using DelimitedFiles
```

```julia
open("mostunstable.dat", "w") do io
    for R = 70 : 5 : 500
        data = readdlm("R=$R.dat",skipstart=3)
        for i = 2 : size(data,1)
            if data[i,5]> data[i-1,5] && data[i,5] > data[i+1,5] && data[i,5] > 0
                mostunstable = data[i,:]
                writedlm(io, [mostunstable])
            end
        end
    end
end
```

```julia
open("Stability_Zarf_Ts=-0.4.dat", "w") do io
    for R = 70 : 5 : 500
        data = readdlm("R=$R.dat",skipstart=3)
        writedlm(io, data)
    end
end
```

```julia
data = readdlm("Stability_Zarf_Ts=-0.4.dat")
```

```julia
for i =  1 : 1 : size(data,1)
    if abs(data[i,2] - 0.0155)<1e-6 && data[i,3] > 0.0 && data[i,5] > 0.0
        data[i,5] *= -1
    end
end
writedlm("Stability_Zarf_Ts=-0.4.dat", data)
```
